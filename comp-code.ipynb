# Comparison Code

This code is used to compare found data values with those in literature. 

Literature texts used: 
Wevers et. al 2017 
Graur et. al 2018 
Van Velzen et. al 2018

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Setting up index

```python
M_BH = []
M_GAL =[]

BHerr_max = []
BHerr_min = []

GALerr_max = []
GALerr_min = []

M_BH_wvrs = []
M_BHerr_max_wvrs = []
M_BHerr_min_wvrs = []

M_GAL_grr = []
M_GALerr_max_grr = []
M_GALerr_min_grr = []

M_GAL_van18 = []
```

```python
with open("Combining_our_masses_incomplete_no_headers_p.txt") as file:
    for line in file:
        data = line.split(',')
        M_BH.append(float(data[4]))
        M_GAL.append(float(data[1]))
        BHerr_max.append(float(data[5]))
        BHerr_min.append(float(data[6]))
        GALerr_max.append(float(data[2]))
        GALerr_min.append(-float(data[3]))
        M_BH_wvrs.append(float(data[7]))
        M_BHerr_max_wvrs.append(float(data[8]))
        M_BHerr_min_wvrs.append(float(data[9]))
        M_GAL_grr.append(float(data[13]))
        M_GALerr_max_grr.append(float(data[14]))
        M_GALerr_min_grr.append(float(data[15]))
        M_GAL_van18.append(float(data[16]))
```

#### Plotting initial scatter graph

```python
yerror = [GALerr_min, GALerr_max]
xerror = [BHerr_min, BHerr_max]

plt.ylabel('Galaxy Mass')
plt.title('A plot of galaxy mass vs black hole mass for our sample')
plt.xlabel('Black Hole Mass')
plt.errorbar(M_BH, M_GAL, yerr = yerror, xerr = xerror, fmt='o')
plt.show()
```
